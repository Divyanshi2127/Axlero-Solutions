````python
import pandas as pd
import joblib
from sqlalchemy import create_engine
````

````python
DATABASE_URL = "postgresql://username:password@localhost:5432/supplyprescript"

engine = create_engine(DATABASE_URL)
````

````python
predictions = pd.read_csv("../optimization/optimized_shipments.csv")

predictions.to_sql(
    "shipment_predictions",
    engine,
    if_exists="replace",
    index=False
)
````

````python
df = pd.read_sql(
    "SELECT * FROM shipment_predictions",
    engine
)

df.head()
````

````python
feedback = pd.DataFrame({

    "Shipment_ID":[1,2,3],

    "Predicted":[1,0,1],

    "Actual":[1,1,0]

})

feedback
````

````python
accuracy = (

    feedback["Predicted"]

    ==

    feedback["Actual"]

).mean()

print("Current Accuracy:",accuracy)
````

````python
if accuracy < 0.90:

    print("Model Drift Detected")

else:

    print("Model Stable")
````
