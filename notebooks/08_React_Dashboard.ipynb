```python
import { Button, TextField } from "@mui/material";
import api from "../services/api";

function PredictionForm(){

    async function predict(){

        const response = await api.post("/predict",{

            Type:0,

            Days_for_shipment_scheduled:3

        });

        console.log(response.data);

    }

    return(

        <>

        <TextField label="Days"/>

        <Button

            onClick={predict}

            variant="contained"

        >

            Predict

        </Button>

        </>

    );

}

export default PredictionForm;
```

```python
import { useEffect,useState } from "react";

import api from "../services/api";

function OptimizationTable(){

    const[data,setData]=useState([]);

    useEffect(()=>{

        api.get("/optimize")

        .then(res=>setData(res.data));

    },[]);

    return(

        <table>

            <thead>

                <tr>

                    <th>Shipment</th>

                    <th>Cost</th>

                    <th>Priority</th>

                </tr>

            </thead>

            <tbody>

            {

                data.map((row,index)=>

                    <tr key={index}>

                        <td>{index+1}</td>

                        <td>{row.Shipping_Cost}</td>

                        <td>{row.Priority}</td>

                    </tr>

                )

            }

            </tbody>

        </table>

    );

}

export default OptimizationTable;
```

```python
import {

BarChart,

Bar,

XAxis,

YAxis,

Tooltip,

CartesianGrid

} from "recharts";

const data=[

{month:"Jan",sales:4200},

{month:"Feb",sales:5100},

{month:"Mar",sales:6000},

{month:"Apr",sales:7200}

];

function Charts(){

return(

<BarChart

width={700}

height={300}

data={data}

>

<CartesianGrid/>

<XAxis dataKey="month"/>

<YAxis/>

<Tooltip/>

<Bar dataKey="sales"/>

</BarChart>

);

}

export default Charts;
```
