# Project Overview

SupplyPrescript is an AI-powered Supply Chain Analytics platform that combines Predictive Analytics and Prescriptive Analytics.

The system predicts shipment delays using machine learning and recommends optimized logistics decisions using Linear Programming.

## Business Problem

Late shipments increase transportation costs, reduce customer satisfaction, and negatively impact supply chain performance.

Traditional reporting systems identify delays after they occur but fail to recommend corrective actions.

SupplyPrescript addresses this challenge by predicting shipment delays and providing optimized shipment recommendations before delays happen.
