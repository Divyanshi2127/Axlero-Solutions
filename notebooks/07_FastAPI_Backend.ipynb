```python
from fastapi import FastAPI
from pydantic import BaseModel
import pandas as pd
import joblib
```

```python
model = joblib.load("../models/best_model.pkl")
```

```python
app = FastAPI(
    title="SupplyPrescript API",
    description="Shipment Delay Prediction API",
    version="1.0"
)
```

```python
class Shipment(BaseModel):

    Type: int
    Days_for_shipment_scheduled: int
    Benefit_per_order: float
    Sales_per_customer: float
    Category_Name: int
    Customer_City: int
    Customer_Country: int
    Customer_Segment: int
    Customer_State: int
    Department_Name: int
    Market: int
    Order_City: int
    Order_Country: int
    Order_Item_Discount: float
    Order_Item_Discount_Rate: float
    Order_Item_Product_Price: float
    Order_Item_Profit_Ratio: float
    Order_Item_Quantity: int
    Sales: float
    Order_Item_Total: float
    Order_Profit_Per_Order: float
    Order_Region: int
    Order_State: int
    Product_Name: int
    Product_Price: float
    Product_Status: int
    Shipping_Mode: int
    Lead_Time: int
    Order_Month: int
    Order_Year: int
    Order_Weekday: int
    Profit_Margin: float
    High_Value_Order: int
```

```python
@app.get("/")
def home():

    return {

        "message": "SupplyPrescript API Running Successfully"

    }
```

```python
@app.get("/model-info")
def model_info():

    return {

        "Model": "XGBoost",

        "Target": "Late Delivery Risk",

        "Version": "1.0"

    }
```

```python
@app.post("/predict")
def predict(data: Shipment):

    df = pd.DataFrame([data.dict()])

    prediction = model.predict(df)[0]

    probability = model.predict_proba(df)[0][1]

    return {

        "Prediction": int(prediction),

        "Probability": float(probability)

    }
```

```python
@app.get("/optimize")
def optimize():

    optimized = pd.read_csv(
        "../optimization/optimized_shipments.csv"
    )

    return optimized.to_dict(orient="records")
```

```python
import uvicorn

uvicorn.run(
    app,
    host="127.0.0.1",
    port=8000
)
```
